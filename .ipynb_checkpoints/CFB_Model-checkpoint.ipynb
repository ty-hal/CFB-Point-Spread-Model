```python
games_api = cfbd.GamesApi(api_config)
# drives_api = cfbd.DrivesApi(api_config)
# plays_api = cfbd.PlaysApi(api_config)
teams_api = cfbd.TeamsApi(api_config)
# conferences_api = cfbd.ConferencesApi(api_config)
# venues_api = cfbd.VenuesApi(api_config)
# coaches_api = cfbd.CoachesApi(api_config)
# players_api = cfbd.PlayersApi(api_config)
rankings_api = cfbd.RankingsApi(api_config)
betting_api = cfbd.BettingApi(api_config)
# recruiting_api = cfbd.RecruitingApi(api_config)
# ratings_api = cfbd.RatingsApi(api_config)
metrics_api = cfbd.MetricsApi(api_config)
stats_api = cfbd.StatsApi(api_config)
```

```python
# games = []
start_year = 2015
end_year = 2023
current_year = 2023

data = pd.DataFrame()
for year in range(start_year, end_year + 1):
    response_games = games_api.get_games(year=year)
    
    # Create a list of dictionaries containing game data
    game_data = [{'id': g.id,
                  'year': g.season,
                  'week': g.week,
                  'neutral_site': g.neutral_site,
                  'home_team': g.home_team,
                  'home_conference': g.home_conference,
                  'home_points': g.home_points,
                  'home_elo': g.home_pregame_elo,
                  'away_team': g.away_team,
                  'away_conference': g.away_conference,
                  'away_points': g.away_points,
                  'away_elo': g.away_pregame_elo} for g in response_games
                    if g.home_conference is not None
                    and g.away_conference is not None
                    and g.home_points is not None
                    and g.away_points is not None
                 ]
    
    # Convert the list of dictionaries to a DataFrame and concatenate with data
    data = pd.concat([data, pd.DataFrame(game_data)], ignore_index=True)

data['margin'] = data['away_points'] - data['home_points']
data.head()
```

```python
betting_data = pd.DataFrame()

for year in range(start_year, end_year + 1):
    response_lines = betting_api.get_lines(year=year)
    year_betting_data = [{'id': g.id,
                      'spread': [l.spread for l in g.lines if l.provider == 'consensus']} 
                    for g in response_lines
                    if g.lines is not None
                    and len(g.lines) != 0
                 ]
    betting_data = pd.concat([betting_data, pd.DataFrame(year_betting_data)], ignore_index = True)

betting_data['spread'] = betting_data['spread'].apply(lambda x: float(x[0]) if x and x[0] is not None else None)
betting_data = betting_data.dropna()

betting_data.head()
```

```python
data = pd.merge(data, betting_data, on='id', how='left')

data.head()
```

```python
# Create a dictionary to collect talent data
talent_dict = {}

# Loop through years and make API calls
for year in range(start_year, end_year + 1):
    talent_for_year = teams_api.get_talent(year=year)
    for tt in talent_for_year:
        key = (tt.school, tt.year)
        if key not in talent_dict:
            talent_dict[key] = tt.talent

# Map the talent data to the DataFrame based on 'school' and 'year'
data['home_talent'] = data.apply(lambda row: talent_dict.get((row['home_team'], row['year']), None), axis=1)
data['away_talent'] = data.apply(lambda row: talent_dict.get((row['away_team'], row['year']), None), axis=1)

data.head()
```

```python
# Create a dictionary to collect talent data
rankings_dict = {}

# Only looking at these polls
unique_poll_names = set(['AP Top 25', 'Coaches Poll'])

# Loop through years and make API calls
for year in range(start_year, end_year + 1):
    rankings_for_year = rankings_api.get_rankings(year=year)
    for rr in rankings_for_year:
        for poll in rr.polls:
            if poll.poll in unique_poll_names:
                for rank in poll.ranks:
                    key = (rank.school, rr.season, rr.week, poll.poll)
                    rankings_dict[key] = rank.points

# Map the ranking data to the DataFrame based on 'home_team', 'year', and 'week'
for poll_name in unique_poll_names:
    home_column_name = f'home_{poll_name.replace(" ", "_").lower()}_points'
    away_column_name = f'away_{poll_name.replace(" ", "_").lower()}_points'
    data[home_column_name] = data.apply(
        lambda row: rankings_dict.get((row['home_team'], row['year'], row['week'], poll_name), 0),
        axis=1
    )
    data[away_column_name] = data.apply(
        lambda row: rankings_dict.get((row['away_team'], row['year'], row['week'], poll_name), 0),
        axis=1
    )
    
data.head()
```
